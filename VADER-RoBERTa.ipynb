```python
from sklearn.metrics import roc_curve, roc_auc_score
import seaborn as sns
import pandas as pd

robertaresults = pd.read_csv("RoBERTa.csv")
vaderresults = pd.read_csv("VADER.csv")


fpr, tpr, thresholds = roc_curve(vaderresults.label, vaderresults.sentiment_score)
vader_auroc = roc_auc_score(vaderresults.label, vaderresults.sentiment_score)
ax=sns.lineplot(x=fpr,y=tpr,label=f"VADER (AUROC) = {vader_auroc:.3f}")

fpr, tpr, thresholds = roc_curve(robertaresults.label, robertaresults.score)
roberta_auroc = roc_auc_score(robertaresults.label, robertaresults.score)
ax=sns.lineplot(x=fpr,y=tpr,label=f"RoBERTa (AUROC) = {roberta_auroc:.3f}")

ax.set(title="VADER vs. RoBERTa ROC Curve Graph", xlabel="fpr", ylabel="tpr")
```
